```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data.head()
```

```python
X = data.drop(columns='Salary',axis=1)
Y = data["Salary"]
```

```python
print(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
regr = LinearRegression()
```

```python
regr.fit(X_train,y_train)
```

## Prediction of Salary of 14 Years of Experience

```python
input_data=(14)

#change input data as numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshape the numpy array 
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction =regr.predict(input_data_reshaped)

print("The Salary of employee after 14 years of experience is :" ,prediction)
```

## Accuracy Score of the System 

```python
regr.score(X_test,y_test)
```
